```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import tensorflow as tf
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('cardio_train1.csv')
data.head()
```

```python
data.info()
```

```python
data.corr()
```

```python
X = data.iloc[:, 1:-1]
y = data.iloc[:, -1]
```

```python
mns = MinMaxScaler()
X = mns.fit_transform(X)
```

```python
X
```

```python
np.shape(X)
```

```python
y =pd.get_dummies(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
model_tf = tf.keras.models.Sequential([
    tf.keras.layers.Dense(11, activation = 'relu', input_shape=(11,)),
    tf.keras.layers.Dense(30, activation = 'relu'),
    tf.keras.layers.Dense(2, activation = 'softmax')
])
```

```python
model_tf.compile(optimizer = tf.keras.optimizers.Adam(0.001),
                 loss = tf.keras.losses.CategoricalCrossentropy(),
                 metrics = [tf.keras.metrics.CategoricalAccuracy()])
```

```python
model_tf.fit(X_train, y_train, epochs = 10, validation_data = (X_test, y_test), batch_size = 700)
```
